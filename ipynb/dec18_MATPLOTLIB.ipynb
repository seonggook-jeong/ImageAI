```python
import matplotlib.pyplot as plt

df = pd.read_excel('./datas/시도별 전출입 인구수.xlsx', fillna = 0, header = 0)

df = df.fillna(method = 'ffill')
```

```python
df
```

```python
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
```

```python
mask
```

```python
df_seoul = df[mask]
```

```python
df_seoul
```

```python
df_seoul = df_seoul.drop(columns = '전출지별')
```

```python
df_seoul
```

```python
df_seoul.rename({'전입지별':'전입지'}, axis = 1, inplace = True)
```

```python
df_seoul
```

```python
df_seoul.set_index('전입지', inplace = True)
```

```python
df_seoul
```

```python
from matplotlib import font_manager, rc
```

```python
sr_one = df_seoul.loc['경기도']

path = 'C:\Windows\Fonts\Batang.ttc'

font_name = font_manager.FontProperties(fname = path).get_name()
rc('font', family = font_name)

plt.figure(figsize = (14, 7))  #그래프 사이즈
plt.plot(sr_one.index, sr_one.values, marker = '*', markersize = 5)  #그래프 그리기(x축, y축 지정)
plt.xticks(rotation = 'vertical')   #x틱스 회전
plt.title('서울->경기도 인구 이동')  #그래프 제목
plt.legend(labels = ['서울->경기'], loc = 'best')     #범례
plt.xlabel('연도')    #x축 라벨
plt.ylabel('이동 인구수')   #y축 라벨
plt.ylim(50000, 800000)
plt.annotate('', xy = (20, 620000), xytext = (2, 290000), xycoords = 'data', arrowprops = dict(arrowstyle = '->', color = 'orange', lw = 3))
plt.annotate('인구이동 증가 (1970~1995)', xy = (10, 360000), rotation = 32, va = 'baseline', ha = 'center', fontsize = 15)
plt.show()  #변경사항 적용
```

```python
plt.style.use('ggplot')   #그래프 스타일 지정
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(1, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(sr_one, 'o', markersize = 5)
ax2.plot(sr_one, 'o', color = 'green', markersize = 5)
ax2.legend(labels = ['서울->경기'], loc = 'best')

ax1.set_ylim(50000, 800000)
ax2.set_ylim(50000, 800000)

ax1.set_xticklabels(sr_one.index, rotation = 75)
ax2.set_xticklabels(sr_one.index, rotation = 75)
plt.show()
```

```python
# 스타일 서식 지정
plt.style.use('ggplot') 

# 그래프 객체 생성 (figure에 1개의 서브 플롯을 생성)
fig = plt.figure(figsize=(20, 5))   
ax = fig.add_subplot(1, 1, 1)

# axe 객체에 plot 함수로 그래프 출력
ax.plot(sr_one, marker='o', markerfacecolor='orange', markersize=10, 
        color='olive', linewidth=2, label='서울 -> 경기')
ax.legend(loc='best')

#y축 범위 지정 (최소값, 최대값)
ax.set_ylim(50000, 800000)

# 차트 제목 추가
ax.set_title('서울 -> 경기 인구 이동', size=20)

# 축이름 추가
ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size = 12)

# 축 눈금 라벨 지정 및 75도 회전
ax.set_xticklabels(sr_one.index, rotation=75)

# 축 눈금 라벨 크기
ax.tick_params(axis="x", labelsize=10)
ax.tick_params(axis="y", labelsize=10)

plt.show()  # 변경사항 저장하고 그래프 출력
```
