```python
def addlut(op):
    lut = [x + op if x + op < 255 else 255 for x in range(256)]
    return lut
```

```python
addlut(5)
```

```python
img2 = cv2.LUT(img, np.array(addlut(100)))
```

```python
plt.imshow(img, cmap = 'gray')
plt.show()
plt.imshow(img2, cmap = 'gray')
plt.show()
```

```python
def binlut(t):
    lut = np.array([0 if x < t else 255 for x in range(256)])
    return lut
```

```python
img2 = cv2.LUT(img, binlut(127))

plt.imshow(img2, cmap = 'gray')
```

```python
def posterizinglut(step):
    t = 256 // step
    lut = [(x//t) * t for x in range(256)]
    return np.array(lut)
```

```python
img2 = cv2.LUT(img, posterizinglut(10))

plt.imshow(img2, cmap = 'gray')
```

```python
def rangelut(t1, t2):
    return np.array([255 if t1 < x < t2 else x for x in range(256)]).astype("uint8")
```

```python
rangelut(200, 230)
```
