실습] ‘titanic’ 데이터셋의 탑승객 데이터를 사용하여 탑승객의 생존 여부를 예측‘survived’ 열에 생존자는 1로, 미생존자는 0으로 표시하여 저장하고 있다.
객실 데크 위치를 나타내는 ‘deck’열에는 유효한 값이 203개로 전체 891명의 승객중 688명의 데이터가 존재하지 않으므로 ‘deck’열은 제거한다.
승선도시를 나타내는 ‘embark_town’열은 ‘embarked’열과 동일한 의미를 갖기 때문에 중복을 없애기 이해 열 자체를 제거한다.
승객의 나이를 나타내는 ‘age’열에 누락된 데이터는 177개 포함되어 있다 분석에 포함시켜야 하는 중요한 속성으로 판단되는 경우 예측 결과에 영향을 최소화하는 경우 평균 나이로 치환하는 방법이 가능하지만, 누락 데이터 행을 제거하고 714명의 승객만을 분석 대상으로 한다.
변수로 사용할 후보열 – survived, pclass, sex, age, sibsp, parch, embarked 


```python
import pandas as pd
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
titanic.info()
```

```python
titanic.drop(['embark_town', 'deck'], axis = 1, inplace = True)
```

```python
titanic.head()
```

```python
titanic.dropna(subset = ['age'], how = 'any', axis = 0, inplace = True)   #NA값이 있는 모든 열 지우기
```

```python
titanic.info()
```

```python
titanic['embarked'].value_counts(dropna = True)    #항목별 빈도수 출력
```

```python
most_freq = titanic['embarked'].value_counts(dropna = True).idxmax()
print(most_freq)
```

```python
titanic['embarked'].fillna(most_freq, inplace = True)
titanic.describe(include = 'all')
```

```python
onehot_sex = pd.get_dummies(titanic['sex'])
```

```python
titanic = pd.concat([titanic, onehot_sex], axis = 1)
```

```python
titanic
```

```python
onehot_embarked = pd.get_dummies(titanic['embarked'], prefix = 'town')
```

```python
titanic = pd.concat([titanic, onehot_embarked], axis = 1)
```

```python
#분석에 활용할 변수 선택
X = titanic[['pclass', 'male', 'female', 'age', 'sibsp', 'parch', 'town_C', 'town_S', 'town_Q']]
Y = titanic['survived']
```

```python
X
```

```python
#KNN 분류 분석
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, Y_train)
```

```python
y_predict = knn.predict(X_test)
```

```python
from sklearn import metrics
knn_matrix = metrics.confusion_matrix(Y_test, y_predict)
print(knn_matrix)
```

```python
knn_report = metrics.classification_report(Y_test, y_predict)
```

```python
print(knn_report)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, y_predict)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(X_train, Y_train)

y_predict = knn.predict(X_test)

from sklearn import metrics
knn_matrix = metrics.confusion_matrix(Y_test, y_predict)
print(knn_matrix)

knn_report = metrics.classification_report(Y_test, y_predict)

print(knn_report)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train, Y_train)

y_predict = knn.predict(X_test)

from sklearn import metrics
knn_matrix = metrics.confusion_matrix(Y_test, y_predict)
print(knn_matrix)

knn_report = metrics.classification_report(Y_test, y_predict)

print(knn_report)
```

# SVM (Support Vector Machine)

```python
from sklearn import svm
svm_model = svm.SVC(kernel = 'rbf')
svm_model.fit(X_train, Y_train)

y_predict = svm_model.predict(X_test)
```

```python
svm_matrix = metrics.confusion_matrix(Y_test, y_predict)
print(svm_matrix)

svm_report = metrics.classification_report(Y_test, y_predict)

print(svm_report)
```

```python
accuracy_score(Y_test, y_predict)
```
