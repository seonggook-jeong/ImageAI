## 1-1. 피보나치 수열

```python
#정의 그대로를 구현하면:
def fibo(n):
    if n < 3:
        return 1
    else: return fibo(n-1) + fibo(n-2)
```

```python
fibo(40)
```

재귀가 너무 깊어져서 너무 오래 걸린다.

```python
def get_fibo(N):
    fibo_list = [0]
    a=0;b=1
    for _ in range(N):
        fibo_list.append(b)
        a, b = b, a+b
    return fibo_list[-1]
```

```python
for i in range(1, 10):
    print(get_fibo(i))
```

```python
get_fibo(200)
```

완전 빠르다.

딕셔너리로 해보라고 하신다.

```python
def get_fibo(N):
    fibo_dict = {}
    a = 0; b = 1
    for k in range(1, N+2):
        fibo_dict[k] = b
        a, b = b, a+b
    return fibo_dict[N]
```

```python
for k in range(1, 100):
    print(get_fibo(k))
```

```python
#솔루션: 중복 연산을 제거하는 방식; 캐시를 활용한다.
pre = {1:1, 2:1}
def fibo2(n):
    v = pre.get(n)
    if v != None:    #이미 계산된 값이 있는 경우
        return v
    else:            #이미 계산된 값이 없는 경우
        sum = fibo2(n-1) + fibo2(n-2)
        pre[n] = sum
        return sum
```

```python
fibo2(100)
```

```python
for k in range(1, 100):
    print(fibo2(k))
```

```python
print(pre)
```

## 2-1. Weather intent에 연결하기

```python
#도시명을 인자로 받아서 날씨를 리턴하는 함수
def getWeather(city):
    url = 'https://search.naver.com/search.naver?query='
    url += urllib.parse.quote_plus(city + ' 날씨')
    #print(url)
    bs = BeautifulSoup(urllib.request.urlopen(url).read(), 'html.parser')
    temp = bs.select("span.todaytemp")
    desc = bs.select("p.cast_txt")
    
    return temp[0].text + '˚ /' + desc[0].text
```

```python
print(getWeather("제주도"))
```

데이터를 조합해서 리턴하는 건 별로 좋지 않다.

개별 데이터를 조작하기 힘들기 때문
- 만드는 사람 따로 갖다 쓰는 사람 따로인 경우가 많기 때문에 이럴 경우 문제가 발생한다.

### =>  딕셔너리로 리턴하는 것이 좋다.

```python
def getWeather(city):
    url = 'https://search.naver.com/search.naver?query='
    url += urllib.parse.quote_plus(city + ' 날씨')
    #print(url)
    bs = BeautifulSoup(urllib.request.urlopen(url).read(), 'html.parser')
    temp = bs.select("span.todaytemp")
    desc = bs.select("p.cast_txt")

    return {"temp": temp[0].text,  "desc": desc[0].text}
```

```python
print(getWeather("제주도"))
print(type(getWeather("제주도")))
```

```python
def get_answer(text, sessionId):
    data_send = {
        'query': text, 'sessionId': sessionId,
        'lang': 'ko', 'timezone' : 'Asia/Seoul'
    }
    data_header = {
        'Authorization': 'Bearer b76b00695b5a4dbca8e7cd27d612b9df',   #API Key
        'Content-Type': 'application/json; charset=utf-8'
    }

    dialogflow_url = 'https://api.dialogflow.com/v1/query?v=20150910'
    res = requests.post(dialogflow_url, data=json.dumps(data_send), headers=data_header)
    
    """
    post 방식 리퀘스트; url, data, headers
        - data: post할 데이터
            - 4개 정보를 갖는다: 'query', 'sessionId', 'lang', 'timezone'
        - headers: API key를 딕셔너리 타입으로 전달한다.
    """
    
    if res.status_code == requests.codes.ok:   #정상적으로 호출되었을 때
        return res.json()    
    return {}                                  #정상적으로 호출되지 않았을 때
```

```python
while True:
    dict = get_answer(input(), "user01")
    answer = dict['result']['fulfillment']['speech']
    print(dict['result']['metadata']['intentName'])

    if dict['result']['metadata']['intentName'] == 'Order2':
        price = {'짜장면': 5000, '짬뽕': 10000, '탕수육': 20000}
        params = dict['result']['parameters']['food_number']

        output = [food.get("number-integer", 1) * price[food['food']] for food in params]
        #get을 쓰는 이유는 음식이 하나인 경우 수량을 명시하지 않는 경우도 있기 때문(에러핸들링의 측면에서)
        print(sum(output))

    elif dict['result']['metadata']['intentName'] == 'Weather' and dict['result']['actionIncomplete'] == False:
        date = dict['result']['parameters']['date']
        geo_city = dict['result']['parameters']['geo-city']
        
        info = getWeather(geo_city)
        
        print(f"{geo_city} 날씨 정보 : {info['temp']} / {info['desc']}")

    else:
        print('Bot: ' + answer)
```

## 4-2. Flask로 간단한 웹 서버 구동하기
- 계속해서 실행되어야하기에 아래의 내용을 별도의 `.py` 파일로 생성해준다.

```python
app
```

```python
from flask import Flask
from flask import request

app = Flask(__name__)

@app.route('/')
#데코레이터임; 데코레이터 아래에 정의된 함수 앞|뒤로 코드를 붙여준다.
def home():
    name = request.args.get('name') 
    item = request.args.get('item')
    #메인 주소 뒤에 파라미터 지정할 수 있다.
    #e.g. http://127.0.0.1:3000/?name=sujin&item=한글
    return "hello " + name + item

@app.route('/abc')
def abc():
    return "test"

@app.route('/welcome/<name>')
def welcome(name):
    return "welcome " + name

if __name__ == '__main__':
    app.run(host = '0.0.0.0', port = 3000, debug = True)
```

해당 파일을 실행하기 위해서는 명령 프롬프트 창에서
- $`cd 해당 디렉토리`

를 입력한다.

## 4-3. 날씨 가져오는 코드 서버에 올리기

아래의 코드를 별도의 `.py` 파일로 만들어준 뒤 명령프롬프트에서 실행해준다.

```python
from flask import Flask
from flask import request
import urllib
from bs4 import BeautifulSoup

app = Flask(__name__)

def getWeather(city):
    url = 'https://search.naver.com/search.naver?query='
    url += urllib.parse.quote_plus(city + ' 날씨')
    bs = BeautifulSoup(urllib.request.urlopen(url).read(), 'html.parser')
    temp = bs.select("span.todaytemp")
    desc = bs.select("p.cast_txt")

    return {"temp": temp[0].text,  "desc": desc[0].text}
    
@app.route('/weather')
def weather():
    city = request.args.get("city")
    info = getWeather(city)
    return info['temp'] + '도   ' + info["desc"]

if __name__ == '__main__':
    app.run(host = '0.0.0.0', port = 3000, debug = True)
```

하지만 이종 플랫폼 간 통신을 구현하는 것이 최종목표이기에 **json** 타입으로 리턴하도록 수정한다.
- flask 내장 메소드인 `jsonify`를 활용한다.

```python
from flask import Flask, request, jsonify   #jsonify 임포트 해주고
from flask import request
import urllib
from bs4 import BeautifulSoup

app = Flask(__name__)

def getWeather(city):
    url = 'https://search.naver.com/search.naver?query='
    url += urllib.parse.quote_plus(city + ' 날씨')
    bs = BeautifulSoup(urllib.request.urlopen(url).read(), 'html.parser')
    temp = bs.select("span.todaytemp")
    desc = bs.select("p.cast_txt")

    return {"temp": temp[0].text,  "desc": desc[0].text}
    
@app.route('/weather')
def weather():
    city = request.args.get("city")
    info = getWeather(city)
    return jsonify(info)    #json으로 바꿔준다

if __name__ == '__main__':
    app.run(host = '0.0.0.0', port = 3000, debug = True)
```

# 5. DialogFlow 연결하기 (to DF)

DialogFlow는 기본적으로 **POST 방식으로 요청을 보내기 때문에**, 응답 함수에 POST 방식으로 요청이 들어올 것임을 명시해주어야 한다.
- 데코레이터 route 정의부에 methods 인자를 추가해주면 된다.
    - e.g. `@app.route('/dialogflow', methods = ['POST'])`
    - methods에 명시된 방식으로 요청이 들어올 때만 함수가 작동한다.

```python
from flask import Flask, request, jsonify
from flask import request
import urllib
from bs4 import BeautifulSoup

app = Flask(__name__)

def getWeather(city):
    url = 'https://search.naver.com/search.naver?query='
    url += urllib.parse.quote_plus(city + ' 날씨')
    bs = BeautifulSoup(urllib.request.urlopen(url).read(), 'html.parser')
    temp = bs.select("span.todaytemp")
    desc = bs.select("p.cast_txt")

    return {"temp": temp[0].text,  "desc": desc[0].text}
    
@app.route('/weather')
def weather():
    city = request.args.get("city")
    info = getWeather(city)
    return jsonify(info)

@app.route('/dialogflow', methods = ['POST', 'GET'])                 #이 부분을 추가해준다!
def dialogflow():
    res = {'fulfillmentText': 'Hello'}       #'fulfillmentText' 키는 DialogFlow에서 규정된 것; 바꿀 수 없다.
    return jsonify(res)

if __name__ == '__main__':
    app.run(host = '0.0.0.0', port = 3000, debug = True)
```
