# 직접 레이블링 해서 학습시키기

## 1. 라벨링 툴 다운로드

    - https://tzutalin.github.io/labelImg/ 에서

        1. 맨 아래 버전(windows_v1.8.0.zip) 다운로드 받기

        2. 압축풀기
            - 압축을 풀면 파일이 2개 있다
                - labelImg(라벨링 프로그램)
                - data 폴더
                    - predefined_classes.txt
                        - 이 파일에 분류할 클래스명을 한줄에 하나씩 써주고 'classes.txt'로 저장한다.

## 2. 이미지 다운로드하기
    - data 폴더에 다운로드 받는다.
    

## 3. 이미지 라벨링하기
    - labelImg을 활용해서 박스를 쳐준다.
        - 대부분의 버튼은 왼쪽에 있는 세로로 된 툴바에 있음
        1. `Open Dir` 버튼을 눌러서 data 폴더(혹은 이미지를 저장한 폴더)를 선택한다.
        2. `Create RectBox` 버튼을 눌러서 오브젝트가 위치한 곳에 박스를 치고 클래스명을 입력해준다.
        3. `Save` 버튼 아래에 있는 `PascalVOC` 버튼을 누르면 `YOLO` 버튼으로 바뀐다. 이 상태에서 `Save` 버튼을 눌러야 표시한 영역이 YOLO에서 사용하는 .txt 형식으로 저장된다.
        4. `Next Image` 버튼으로 모든 이미지에 대해서 1~3을 수행한다.


## 4. 초기 설정
> 학습시킬 환경에 맞게 각종 파일을 수정해준다.

(각종 파일은 선생님이 올려주신 zip 파일에 모두 포함되어 있었으나, 인터넷에서 다 다운로드 받을 수 있음)

    - darknet.data
        - classes = 클래스 수
        - train = 트레인 이미지 경로(절대경로)들이 한줄에 하나씩 저장된 텍스트파일
        - valid = 트레인 이미지 경로(절대경로)들이 한줄에 하나씩 저장된 텍스트파일
        - names = 클래스명이 한줄에 하나씩 저장된 names 파일
        - backup = 백업 폴더 경로(상대경로)
    - darknet-yolo3.cfg
        - yolo 레이어
            - classes = 클래스 수
        - yolo 레이어의 전 레이어
            - filters =  3 * (5+classes)
                - 여기서 클래스 수에 5를 더해주는 이유: 바운딩박스의 x, y, w, h, p
                
## 5. PC에서 학습
아나콘다 프롬프트를 열어서 `cd snowman` (혹은 위의 파일들이 모두 저장된 디렉토리)으로 이동한다.

`darknet_no_gpu detector train darknet.data darknet-yolov3.cfg darknet53.conv.74`

를 입력하면 학습이 이루어진다.

## 5-1. Linux에서 학습시키기

### 1. Putty로 jupyter notebook 실행

`conda activate`

`jupyter notebook --ip 0.0.0.0 --port 8895`

브라우저 주소창에 `ip 주소:포트번호`를 입력하면 jupyter notebook을 실행할 수 있다.
    

### 2. snowman.zip (혹은 위에서 생성한 파일들이 모두 들어있는 zip 파일)을 업로드


### 3. 새로운 Putty 실행해서 snowman.zip 압축 풀기

`unzip snowman.zip -d snowman`
    
#### 이제 보다 쉽게 편집이 가능하다.

### 4. 몇몇 파일을 Linux 환경에 맞게 수정해준다.
> 경로명이 들어가는 파일은 모두 바꿔줘야한다.
    - darknet.data
    - test.txt
    - train.txt
    
    의 경로들을 모두 Linux 가상환경의 **절대경로**로 바꿔준다.
    
#### 이미지 이름에 image.jpg가 있으면 안된다.
--> yolo가 image를 만나면 label.txt를 예상하기 때문
    
### 5. 학습 실행하기

새로운 Putty 창을 열어준다.

#### 만약 Makefile을 수정했다면 `make` 명령어로 새로 컴파일 해주어야 한다.

`cd darknet`
    
`./darknet detector train ~/snowman/snowman/darknet.data ~/snowman/snowman/darknet-yolov3.cfg ~/snowman/snowman/darknet53.conv.74 > ~/snowman/snowman/snowman.log`
- linux 가상환경에 맞춰서 경로를 써준다.
- > 뒷부분은 로그정보를 저장할 경로를 지정해주는 것이다.
    - 이렇게 하면 프롬프트 창에는 로그정보가 출력되지 않는다.
    
### 6. 로그 확인하기
새로운 Putty 창에서 `grep "avg" snowman/snowman/snowman.log`을 입력해서 로그정보를 확인할 수 있다.
- `grep "키워드" 로그파일명`
     - "키워드"가 포함된 로그 정보만 보여준다.
