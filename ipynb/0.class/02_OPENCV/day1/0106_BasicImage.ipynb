### 컬러 출력

```python
r = np.array([
    [255, 255, 255, 255],
    [0, 0, 0, 0],
    [0, 0, 0, 0],
    [255, 255, 255, 255]
], dtype = 'uint8')

g = np.array([
    [0, 0, 0, 0],
    [255, 255, 255, 255],
    [0, 0, 0, 0],
    [255, 255, 255, 255]
], dtype = 'uint8')

b = np.array([
    [0, 0, 0, 0],
    [0, 0, 0, 0],
    [255, 255, 255, 255],
    [0, 0, 0, 0]
], dtype = 'uint8')

colors = np.dstack([r, g, b])    #dstack == depthstack; z축으로 붙임
print(colors.shape)
plt.imshow(colors)
```

### 컬러 출력 예제

```python
r = np.array([
    [255, 255, 255, 255],
    [0, 0, 0, 0],
    [0, 0, 0, 0],
    [255, 255, 255, 255]
], dtype = 'uint8')
g = np.zeros((4, 4), dtype = 'uint8')
b = np.zeros((4, 4), dtype = 'uint8')

colors = np.dstack([r, g, b])    #dstack == depthstack; z축으로 붙임
plt.imshow(colors)
plt.show()

r = np.zeros((4, 4), dtype = 'uint8')
g = np.array([
    [0, 0, 0, 0],
    [255, 255, 255, 255],
    [0, 0, 0, 0],
    [255, 255, 255, 255]
], dtype = 'uint8')
b = np.zeros((4, 4), dtype = 'uint8')

colors = np.dstack([r, g, b])    #dstack == depthstack; z축으로 붙임
plt.imshow(colors)
plt.show()

r = np.zeros((4, 4), dtype = 'uint8')
g = np.zeros((4, 4), dtype = 'uint8')
b = np.array([
    [0, 0, 0, 0],
    [127, 127, 127, 127],
    [200, 200, 200, 200],
    [255, 255, 255, 255]
], dtype = 'uint8')

colors = np.dstack([r, g, b])    #dstack == depthstack; z축으로 붙임
plt.imshow(colors)
plt.show()
```

### 피부색 디텍션

```python
img = cv2.imread('iu.jpg')
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h = hsv[:, :, 0]
s = hsv[:, :, 1]
v = hsv[:, :, 2]
```

#### 1. H값 만 가지고 거르기

```python
result = img.copy()
```

```python
for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        if h[r, c] >= 0 and h[r, c] <= 20:   #피부색일 경우
            result[r, c, :] = img[r, c, :]   #원본에서 가져옴

        else:                                #피부색이 아닐 경우
            result[r, c, :] = 0              #검정색으로 표현

#lower = np.array([0, 48, 80], dtype = 'uint8')
#upper = np.array([20, 255, 255], dtype = 'uint8')

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
```

#### 2. H, S, V 값 모두를 사용하여 거르기

```python
lower = np.array([0, 48, 80], dtype = 'uint8')
upper = np.array([20, 255, 255], dtype = 'uint8')

for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        if (lower <= hsv[r, c]).all() and (hsv[r, c] <= upper).all():   #피부색일 경우
            result[r, c, :] = img[r, c, :]   #원본에서 가져옴

        else:                                #피부색이 아닐 경우
            result[r, c, :] = 0              #검정색으로 표현


plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
```

```python
#솔루션
lower = np.array([0, 48, 80], dtype = 'uint8')
upper = np.array([20, 255, 255], dtype = 'uint8')

for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        bH = h[r, c] >= 0 and h[r, c] <= 20
        bS = s[r, c] >= 48 and s[r, c] <= 255
        bV = v[r, c] >= 80 and v[r, c] <= 255
        if bH and bS and bV:   #피부색일 경우
            result[r, c, :] = img[r, c, :]   #원본에서 가져옴

        else:                                #피부색이 아닐 경우
            result[r, c, :] = 0              #검정색으로 표현


plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
```

#### 3. 마스크 생성하기

```python
mask = img.copy()

for r in range(h.shape[0]):
    for c in range(h.shape[1]):
        bH = h[r, c] >= 0 and h[r, c] <= 20
        bS = s[r, c] >= 48 and s[r, c] <= 255
        bV = v[r, c] >= 80 and v[r, c] <= 255
        if bH and bS and bV:   #피부색일 경우
            mask[r, c, :] = 255   

        else:                  #피부색이 아닐 경우
            mask[r, c, :] = 0     #검정색으로 표현


plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))
```

## 히스토그램 그리기

```python
train = cv2.imread("patch.png")
print(train.shape)
```

```python
b, g, r = cv2.split(train)    #split: 채널별로 분리; 위의 슬라이싱보다 훨씬 빠름
```

```python
plt.hist(b.ravel(), 256, [0, 256], color = 'b')
plt.hist(g.ravel(), 256, [0, 256], color = 'g')
plt.hist(r.ravel(), 256, [0, 256], color = 'r')
plt.show()
```

```python
#HSV로도 해보기
h, s, i = cv2.split(cv2.cvtColor(train, cv2.COLOR_BGR2HSV))

plt.hist(h.ravel(), 256, [0, 256], color = 'b')
plt.hist(s.ravel(), 256, [0, 256], color = 'g')
plt.hist(i.ravel(), 256, [0, 256], color = 'r')
plt.show()
```
