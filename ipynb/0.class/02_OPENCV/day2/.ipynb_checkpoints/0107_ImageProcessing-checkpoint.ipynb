## 1-2. 정규분포와 확률을 이용한 크로마키

```python
#학습
train = cv2.imread('patch3.png')

b, g, r = cv2.split(train)

#채널별 최댓값을 뽑아내기 위해 히스토그램을 변수에 저장
hb = plt.hist(b.ravel(), 256, [0, 256], color = 'b')
hg = plt.hist(g.ravel(), 256, [0, 256], color = 'g')
hr = plt.hist(r.ravel(), 256, [0, 256], color = 'r')

maxb = np.where(hb[0] == np.max(hb[0]))[0][0]
maxg = np.where(hg[0] == np.max(hg[0]))[0][0]
maxr = np.where(hr[0] == np.max(hr[0]))[0][0]
plt.show()

print(maxb)
print(maxg)
print(maxr)

print('\n')

print(np.mean(b))
print(np.mean(g))
print(np.mean(r))

t = 7
p = sp.norm(0, 1).pdf((9-t)/4)
print(p)
```

```python
#테스트
test = cv2.imread("zoo.jpg")

for r in range(test.shape[0]):
    for c in range(test.shape[1]):
        if np.array_equal(test[r, c, :], [maxb, maxg, maxr]):
            test[r, c, :] = 255
plt.imshow(cv2.cvtColor(test, cv2.COLOR_BGR2RGB))
plt.show()
```

## LUT(Look Up Table)
- 산술연산 프로세싱보다 LUT 참조하는 것이 더 빠르다.
    - 산술연산은 이미지 사이즈에 따라 처리 속도가 달라짐
    - LUT 참조는 **이미지 사이즈에 상관없이** 256번만 돌아감 --> 효율적
    
- LUT 적용은 OpenCV의 `LUT` 메소드를 활용한다.
    - `이미지2 = cv2.LUT(이미지, LUT테이블)`
    - **주의**: LUT는 넘파이 어레이 타입이어야 한다!!!

## 3-1. 널 변환 LUT

```python
lut1 = [x for x in range(10)]
print(lut1)

lut2 = [x+5 for x in range(10)]
print(lut2)

lut3 = [x+5 if (x+5 <= 9) else 9 for x in range(10)]
print(lut3)

#lut3 = np.clip(lut2, 5, 9)
#print(lut3)

def addlut(op):
    return [x+op if (x+op < 255) else 255 for x in range(256)]

def mullut(op):
    return [x*op if (x*op < 255) else 255 for x in range(256)]

print(mullut(1.5))
```

```python
src2 = cv2.LUT(img, mullut(2.0))
print(src2)
plt.imshow(src2, cmap = 'gray')
```

LUT는 리스트면 안된다. 넘파이!!!!!!!!!!!

```python
#리턴값을 넘파이 어레이로, 데이터 타입을 uint8로 변환해준다.
def mullut(op):
    return np.array([x*op if (x*op < 255) else 255 for x in range(256)]).astype("uint8")
```

```python
src2 = cv2.LUT(img, mullut(1.2))
#print(src2)
plt.imshow(src2, cmap = 'gray')
plt.show()
src2 = cv2.LUT(img, mullut(2.0))
#print(src2)
plt.imshow(src2, cmap = 'gray')
plt.show()

all = np.hstack([cv2.LUT(img, mullut(0.7)), cv2.LUT(img, mullut(1.2)), cv2.LUT(img, mullut(2.0))])
plt.imshow(all, cmap = 'gray')
plt.show()
```

## 3-2. 이진화 LUT

```python
def binlut(t):
    return np.array([0 if x < t else 255 for x in range(256)]).astype("uint8")
```

```python
src3 = cv2.LUT(img, binlut(127))
all = np.hstack([img, src3])
plt.imshow(all, cmap = 'gray')
```

## 3-3. 인버스 LUT

```python
def inverselut():
    return np.array([255-x for x in range(256)]).astype('uint8')
```

```python
src3 = cv2.LUT(img, inverselut())
all = np.hstack([img, src3])
plt.imshow(all, cmap = 'gray')
```

## 3-4. 포스터라이징 LUT
- 포스터라이징의 경우 HSV 채널 중 H 채널만 건드려야 함. 그래야 채도, 명도는 유지되고 색상만 비슷한 색끼리 모을 수 있음.

```python
def posterizinglut(t):
    return np.array([(x//t)*t for x in range(256)]).astype('uint8')
```

```python
src3 = cv2.LUT(img, posterizinglut(72))
all = np.hstack([img, src3])
plt.imshow(all, cmap = 'gray')
```

## 3-5. 범위강조 LUT

```python
def rangelut(t1, t2):
    return np.array([255 if t1 < x < t2 else x for x in range(256)]).astype("uint8")
```

```python
src3 = cv2.LUT(img, rangelut(200, 230))
all = np.hstack([img, src3])
plt.imshow(all, cmap = 'gray')
```

## 3-6. 감마 보정 LUT

```python
def gammalut(gamma):
    invGamma = 1.0 / gamma
    return np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)])
```

```python
src3 = cv2.LUT(img, gammalut(2.2))
src4 = cv2.LUT(img, gammalut(0.5))
plt.imshow(np.hstack([img, src3, src4]), cmap = 'gray')
```

## 부록) 이미지에 텍스트 쓰기

`.putText()`메소드를 통해 이미지에 글씨를 쓸 수 있다.
- cv2.putText(이미지, 글, 위치, 폰트, 크기, 색(RGB), 굵기)

```python
img = cv2.imread('Lena.png', cv2.IMREAD_GRAYSCALE)
print(img.shape)    #그레이스케일로 읽어왔기 때문에 2차원
col = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
print(col.shape)    #BGR로 다시 바꿨기 때문에 3차원

cv2.putText(col, "gamma=0.2", (100, 30),   #이미지, 글씨, 위치좌표
           cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 3)   #폰트, 크기, 색, 굵기

plt.imshow(col)
```

## 6. 감마 컬렉션

```python
#하드 코딩
img = cv2.imread('gamma.jpg')
col = cv2.LUT(img, gammalut(0.5))
cv2.putText(col, "gamma=0.5", (100, 30),
           cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 3)
cv2.imwrite("g1.jpg", col)

col = cv2.LUT(img, gammalut(1.5))
cv2.putText(col, "gamma=1.5", (100, 30),
           cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 3)
cv2.imwrite("g2.jpg", col)

col = cv2.LUT(img, gammalut(2.5))
cv2.putText(col, "gamma=2.5", (100, 30),
           cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 3)
cv2.imwrite("g3.jpg", col)
```

```python
#for문으로
gm = [0.5, 1.2, 2.5, 3.0]
for i in range(len(gm)):
    col = cv2.LUT(img, gammalut(gm[i]))
    cv2.putText(col, f"gamma={i}", (100, 30),
               cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 0), 3)
    cv2.imwrite(f"g{i+1}.jpg", np.hstack([img, col]))
```

6-1 GIF 생성(PIL 라이브러리)
- OpenCV가 영상처리를 위한 라이브러리라면 PIL은 이미지 자체를 다루기 위한 라이브러리
- `.save()`메소드를 통해 gif 파일을 생성할 수 있다.
- 기준이미지.save('파일명.gif', save_all = True, append_images = [더할 이미지들], duration = 지속할 초 * 1000, loop = 반복 횟수(0: 무한))

```python
from PIL import Image
im1 = Image.open('g1.jpg')
im2 = Image.open('g2.jpg')
im3 = Image.open('g3.jpg')
im4 = Image.open('g4.jpg')

im1.save("out3.gif", save_all = True, append_images = [im2, im3, im4], duration = 500, loop = 0)
```
