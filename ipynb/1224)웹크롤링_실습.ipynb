정적 웹 페이지 크롤링과 스크래핑 실습
https://movie.naver.com/movie/point/af/list.nhn?page=1

스크래핑 내용
- 영화 제목 : movie의 콘텐츠
- 평점 : td.title > div > em의 콘텐츠
- 댓글 : td.title의 콘텐츠 중  7번째 자식 콘텐츠

```python
from urllib.request import urlopen
import requests
```

```python
from bs4 import BeautifulSoup
```

### 1. urllib 쓰기

```python
url = 'https://movie.naver.com/movie/point/af/list.nhn?page=1'

html = urlopen(url)
```

```python
html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.title
```

### 1-1 requests 쓰기

```python
r = requests.get(url)
html = r.text

soup = BeautifulSoup(html, 'html.parser')
```

```python
titles = soup.select('.movie')
```

```python
titles
```

```python
points = soup.select('td.title > div > em')
reviews = soup.select('td.title')
```

```python
movie_title = []
movie_point = []
movie_review = []

for dom in titles:
    movie_title.append(dom.text)
    
for dom in points:
    movie_point.append(dom.text)
    
for dom in reviews:
    movie_review.append(dom.contents[6].strip())
```

```python
movie_review
```

```python
for i in range(len(movie_title)):
    print('영화제목:', movie_title[i])
    print('평점:', movie_point[i])
    print('리뷰:', movie_review[i])
    print('--------------------------------------------------------------------------------------')
```
